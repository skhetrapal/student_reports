```python
import matplotlib.pyplot as plt
import numpy as np

# Logistic map function
def logistic(r, x):
    return r * x * (1 - x)

# Bifurcation diagram function
def bifurcation_diagram(r_min, r_max, x0, n, m, p):
    # For each value of r, do the following
    
    # Initialize lists for storing values of r and x
    r_vals = []
    x_vals = []

    # Iterate over values of r
    for r in np.linspace(r_min, r_max, n):
        # Initialize the initial condition for x
        x = x0

        # 1st for loop (To arrive at the converged values of x (If they exist for this particular 'r'))
        for i in range(m):
            x = logistic(r, x)

        # 2nd for loop (Essential as the first few steps of the iter. procedure never converges)
        for k in range(m, m+p):
            x = logistic(r, x)
            r_vals.append(r)
            x_vals.append(x)

    # Plot the bifurcation diagram
    plt.scatter(r_vals, x_vals, s=0.0005, marker='.', color='black')
    plt.xlabel('r')
    plt.ylabel('x')
    plt.show()

# Set the parameters
r_min = 0
r_max = 4
x0 = 0.5
n = 1000
m = 1000
p = 500

# Generate the bifurcation diagram
bifurcation_diagram(r_min, r_max, x0, n, m, p)
```
